```python
# importar librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Guarda los datos en un Data Frame llamdo dfCRE del archivo csv COLOMBIANOS_REGISTRADOS_EN_EL_EXTERIOR_20250403
df = pd.read_csv("COLOMBIANOS_EXTERIOR_LIMPIEZA.csv")
```

```python
# Muestra la cantidad de filas y columnas del data set
df.shape
```

```python
# Metodo value_counts muestra los valores presentes en la columna GENERO realiza el conteo de los valores de las ocurrencias
# ademas se genera otra columna CANTIDAD en frecuencias_genero.columns para las ocurrencias
frecuencia_genero = df['GENERO'].value_counts().reset_index()
frecuencia_genero.columns = ['GENERO', 'CANTIDAD']
print(frecuencia_genero)
```

```python
plt.figure(figsize=(8, 8))  # Tamaño del gráfico
plt.pie(frecuencia_genero['CANTIDAD'],       # Valores numéricos
        labels=frecuencia_genero['GENERO'],  # Etiquetas
        autopct='%1.1f%%',    # Mostrar porcentaje con 1 decimal
        startangle=45)        # Rotar inicio de la gráfica

plt.title('Distribución por genero Colombianos registrados en el exterior')
plt.axis('equal')  # Mantener la forma circular
plt.show()

#Interpretacion
#El genero femenino de Colombianos registrados en el exterior supera el 50% del censo 
```
