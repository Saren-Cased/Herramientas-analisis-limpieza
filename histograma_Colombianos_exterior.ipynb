```python
# histograma es una gr√°fica que muestra las distribuciones de frecuencia
# importar librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Guarda los datos en un Data Frame llamdo dfCRE del archivo csv COLOMBIANOS_REGISTRADOS_EN_EL_EXTERIOR_20250403
df = pd.read_csv("COLOMBIANOS_EXTERIOR_LIMPIEZA.csv")
```

```python
# Muestra la cantidad de filas y columnas del data set
df.shape
```

```python
# Metodo value_counts muestra los valores presentes en la columna EDAD_ANOS realiza el conteo de los 
# valores con mas ocurrencias
df.EDAD_ANOS.value_counts().head(20)
```

```python
# Toma el numero de edad de a√±os por rangos
valores_EDAD_ANOS = df.EDAD_ANOS.value_counts()

# Valores en eje x son visibles
plt.xticks(valores_EDAD_ANOS)

# Funci√≥n formatear los valores de miles en los separadores realiza el cambio de , por .
formatter = FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.'))

# Formato en el eje y
plt.gca().yaxis.set_major_formatter(formatter)

# Adicionar etiquetas al titulo
plt.title('Histograma Colombianos en el exterior por edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia de las edades')

# Filtro para excluir filas
# df = df[~df.EDAD_ANOS.isin(['1-20', '80-120' ])]
# df = df[df.GENERO.isin(['MASCULINO'])]

# Crea el histograma con la columna EDAD_ANOS
n, bins, patches = plt.hist(df.EDAD_ANOS, color="blue")

# Adicion de etiquetas de frecuencia por encima de cada barra, en formato de miles
for i in range(len(n)):
    plt.text(bins[i] + 0.5, n[i] + 0.1, f'{int(n[i]):,}'.replace(',', '.'), ha='center', va='bottom', fontsize=10)

# mostrar histograma
plt.show()

# Interpretacion
# En el histograma de edades de Colombianos en el exterior la tendencia es de 35 a√±os, esta dentro de la barra de frecuencia
# 541022. Las barras de frecuencia 541022, 462461 y 301361 cubre el rango de edades de los 27 a los 59 a√±os, en la barra de
# frecuencia de 462461 se encuentra la edad promedio de 43 a√±os.
```
