```python
# importar librerias necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Guarda los datos en un Data Frame llamdo dfCRE del archivo csv COLOMBIANOS_REGISTRADOS_EN_EL_EXTERIOR_20250403
df = pd.read_csv("COLOMBIANOS_EXTERIOR_LIMPIEZA.csv")
```

```python
# Muestra la cantidad de filas y columnas del data set
df.shape
```

```python
#Metodo describe muestra una descripcion estadistica de la columna indicada
df.GENERO.describe()
```

```python
df[df['GENERO'] == 'FEMENINO'][['GENERO', "EDAD_ANOS"]].value_counts()
```

```python
# Metodo value_counts muestra los valores presentes en la columna GENERO realiza el conteo de los valores de las ocurrencias
# ademas se genera otra columna CANTIDAD en frecuencias_genero.columns para las ocurrencias 
frecuencias_genero = df['GENERO'].value_counts().reset_index()
frecuencias_genero.columns = ['GENERO', 'CANTIDAD']
print(frecuencias_genero)
```

```python
# Crea un gráfico de barras donde las barras muestran la suma total de los GENEROS por cada valor de "GENERO"
sns.barplot(data=frecuencias_genero, x='GENERO', y='CANTIDAD', hue='GENERO', palette='pastel', legend=False)

plt.title('Distribución de Género')
plt.xlabel('GENERO')
plt.ylabel('CANTIDAD')
plt.show()

# Interpretacion
# El genero femenino de Colombianos registrados en el exterior es mayor al de los hombres 
```

```python
# Metodo value_counts muestra los valores presentes en la columna GRUPO_EDAD realiza el conteo de los valores de las ocurrencias
# ademas se genera otra columna CANTIDAD en frecuencias_genero.columns para las ocurrencias 
frecuencias_GRUPO_EDAD = df['GRUPO_EDAD'].value_counts().reset_index()
frecuencias_GRUPO_EDAD.columns = ['GRUPO_EDAD', 'CANTIDAD']
print(frecuencias_GRUPO_EDAD)
```

```python
# Crea un gráfico de barras donde las barras muestran la suma total de los GRUPO_EDAD por cada valor de "GRUPO_EDAD" ademas la
# dimension del tablero se genera el tamaño
plt.figure(figsize=(12, 6))

sns.barplot(data=frecuencias_GRUPO_EDAD, x='GRUPO_EDAD', y='CANTIDAD', hue='GRUPO_EDAD', palette='pastel', legend=False)

plt.title('Distribución grupos de edades Colombianos registrados en el exterior')
plt.xlabel('GRUPO EDAD')
plt.ylabel('CANTIDAD')
plt.show()

# Interpretacion
# Los grupos de edades de Colombianos registrados en el exterior con mayor censo son adulto, adulto mayor y adulto joven. 
```

```python
# Metodo value_counts muestra los valores presentes en la columna NIVEL_ACADEMICO realiza el conteo de los valores de las ocurrencias
# ademas se genera otra columna CANTIDAD en frecuencias_ACADEMICO.columns para las ocurrencias 
frecuencias_ACADEMICO = df['NIVEL_ACADEMICO'].value_counts().reset_index()
frecuencias_ACADEMICO.columns = ['NIVEL_ACADEMICO', 'CANTIDAD']
print(frecuencias_ACADEMICO)
```

```python
# Crea un gráfico de barras donde las barras muestran la suma total de NIVEL_ACADEMICO por cada valor de "NIVEL_ACADEMICO" ademas la
# dimension del tablero se genera el tamaño
plt.figure(figsize=(35, 25))

sns.barplot(data=frecuencias_ACADEMICO, x='NIVEL_ACADEMICO', y='CANTIDAD', hue='NIVEL_ACADEMICO', palette='pastel', legend=False)

plt.title('Distribución nivel academico Colombianos registrados en el exterior')
plt.xlabel('NIVEL ACADEMICO')
plt.ylabel('CANTIDAD')
plt.show()

# Interpretacion
# El nivel academico de Colombianos registrados en el exterior con mayor censo son: no indica, bachillerato y pregado profesional.
```
